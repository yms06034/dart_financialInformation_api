```python
import requests
import pandas as pd

def fetch_and_process_data(year, code, result):
    params = {
        "crtfc_key": KEY,
        "corp_code": result,
        "bsns_year": year,
        "reprt_code": code
    }
    
    response = requests.get(mu_url, params=params)

    if response.status_code == 200:
        try:
            data = response.json()
            if data['status'] == '000':
                df = pd.DataFrame(data['list'])
                df.drop(['rcept_no', 'reprt_code', 'fs_div', 'fs_nm', 'sj_div', 'ord', 'currency'], axis=1, inplace=True)
                df['thstrm_amount'] = pd.to_numeric(df['thstrm_amount'].str.replace(',', ''), errors='coerce')
                df.insert(8, '월평균(백만원)', df['thstrm_amount'] // (12 if code == '11011' else 3 if code == '11013' else 6))
                
                # Dynamically set column names based on the DataFrame's shape
                if df.shape[1] == 15:
                    df.columns = ["사업 연도", "회사코드", "종목코드", "재무제표명", "계정명", "당기명", "당기일자", "당기금액", "월평균(백만원)", "전기명", "전기일자", "전기금액", "전전기명", "전전기일자", "전전기금액"]
                elif df.shape[1] == 14:
                    df.columns = ["사업 연도", "회사코드", "종목코드", "재무제표명", "계정명", "당기명", "당기일자", "당기금액", "월평균(백만원)", "전기명", "전기일자", "전기금액", "당기누적금액", "전기누적금액"]
                else:
                    print(f"Unsupported number of columns: {df.shape[1]}")
                    df = pd.DataFrame()
                    
                return df
        except json.JSONDecodeError as e:
            print("JSON 디코딩 오류:", e)
    else:
        print("API 요청에 실패하였습니다. 상태 코드:", response.status_code)
    return pd.DataFrame()


reprt_codes = ["11011", "11012", "11013", "11014"]
years = ["2022", "2023"] 
result_data = {}  
mu_url = "https://opendart.fss.or.kr/api/fnlttMultiAcnt.json"

for year in years:
    for code in reprt_codes:
        for re in result:
            df = fetch_and_process_data(year, code, re)
            if not df.empty:
                result_data.setdefault(year, {}).setdefault(code, {})[re] = df

dataframes_23 = [result_data['2023'][code][re] for code in reprt_codes for re in result_data['2023'][code] if not result_data['2023'][code][re].empty]
combined_23_df = pd.concat(dataframes_23, axis=0)
combined_23_df.reset_index(drop=True, inplace=True)

dataframes_22 = [result_data['2022'][code][re] for code in reprt_codes for re in result_data['2022'][code] if not result_data['2022'][code][re].empty]
combined_22_df = pd.concat(dataframes_22, axis=0)
combined_22_df.reset_index(drop=True, inplace=True)

all_combined_dataframes = [combined_22_df, combined_23_df]
all_combined_df = pd.concat(all_combined_dataframes, axis=0)
all_combined_df.reset_index(drop=True, inplace=True)

dfs_11011 = [result_data['2023']['11011'][re] for re in result_data['2023']['11011'] if not result_data['2023']['11011'][re].empty] + [result_data['2022']['11011'][re] for re in result_data['2022']['11011'] if not result_data['2022']['11011'][re].empty]
all_combined_11011_df = pd.concat(dfs_11011, axis=0)
all_combined_11011_df.reset_index(drop=True, inplace=True)
```
